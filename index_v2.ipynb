## Beispiel 1: Einfacher Plot


Nehmen wir an, wie wollen die Funktion

\begin{align*}
y=\sin(x) + \cos(x) &&\mbox{für } &&x\in [0;10\pi)&&&&&&
\end{align*}
plotten.

Im ersten Schritt importieren wir die benötigten Bibliotheken:

```python
import numpy as np

import matplotlib.pyplot as plt

%matplotlib inline
```

dann erzeugen wir $x$-Werte

```python
x = np.linspace(0, 10*np.pi, 10**3);
```

und nun die zugehörigen $y$-Werte, d.h. die Funktionswerte

```python
y = np.sin(x) + np.cos(x); 
```

und schon können wir den Plot erstellen:

```python
plt.plot(x,y);
plt.title('mein erster Plot');
```

## Beispiel 2: interaktiver Plot



Jetzt wollen wir den Einfluss von Parametern beobachten und betrachten:

\begin{align*}
  y= \sin(a*x) + \cos(x)  &&\mbox{für } &&x\in [0;10\pi)&&&&&&
\end{align*}



### Funktionsdefinition

Um die unterschiedlichen Ansätze zu veranschaulichen, definieren wir 

```python
def myplot(a): 
    x = np.linspace(0, 10*np.pi, 10**3);
    y = np.sin(a*x) + np.cos(x); 
    plt.plot(x,y); 
    plt.show();
```

###  Standard Plot

Wenn wir den Wert von $a$ in der unteren Zelle verändern, abspeichern und die Zelle durchlaufen lassen, verändert sich die Grafik.

```python
a=1 ;
myplot(a)
```

### Interaktiver Plot



Schöner ist es allerdings einen Slider zu erzeugen und den Einfluss des Parameters direkt mitverfolgen zu können. Dazu benötigen wir die _interactive_-Funktion aus der ipywidgets-Bibliothek.

```python
from ipywidgets import interactive
```

Jetzt können wir einen interaktiven Plot erstellen

```python
interactive_plot=interactive(myplot, a = (-20,20, 0.1));
interactive_plot
```
