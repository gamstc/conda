## Beispiel: Plot erstellen

Wir wollen die Funktion
\begin{align*}
y=\sin(x) + \cos(x) &&\mbox{für } &&x\in [0;10\pi)&&&&&&
\end{align*}
plotten.

Im ersten Schritt importieren wir die benötigten Bibliotheken:

```python
import numpy as np

import matplotlib.pyplot as plt

%matplotlib inline
```

jetzt definieren wir die Funktion, die wir plotten möchten

```python
a = 1;
x = np.linspace(0, 10*np.pi, 10**3);
y = np.sin(a*x) + np.cos(x); 
```

und nun noch der Plot-Befehl:

```python
plt.plot(x,y);
plt.title('mein erster Plot');
```
